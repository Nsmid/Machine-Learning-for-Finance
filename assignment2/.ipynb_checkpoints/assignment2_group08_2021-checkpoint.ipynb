Explain what a k-means model is.

- K Means algorithm cluster data and try to separate it into the  optimal number of clusters with similar variance, with minimising different criteria as inertia or other. We need to specify how many clusters we want from the input data frame.The K-means algorithm aims to choose centroids that minimise the inertia, or other criteria.

Estimate a k-means model with 3 clusters for the variables `LoanAmount`, `Zip`, `ProbabilityDefault` and `AgeLoan`.

````python
df_data_kmeans=df_data[['LoanAmount', 'Zip', 'ProbabilityDefault', 'AgeLoan']]
````

````python
Kmean = MiniBatchKMeans(n_clusters=3, random_state=25)
````

````python
Kmean.fit(df_data_kmeans)
````

````python
Kmean.cluster_centers_
````

````python
y_kmeans=Kmean.predict(df_data_kmeans)
````

Determine with the elbow method with inertia what the right number of clusters is. Please plot the inertia and explain your choice of clusters.

````python
inertia_list = []
for num_clusters in range(1, 10):
    kmeans_model = KMeans(n_clusters=num_clusters, init="k-means++")
    kmeans_model.fit(df_data_kmeans)
    inertia_list.append(kmeans_model.inertia_)
````

````python
plt.figure(figsize=(10,7))
plt.plot(range(1,10),inertia_list)
plt.scatter(range(1,10),inertia_list)
plt.scatter(3, inertia_list[3], marker="X", s=300, c="r")
plt.xlabel("Number of Clusters", size=13)
plt.ylabel("Inertia Value", size=13)
plt.title("Different Inertia Values for Different Number of Clusters", size=15)
````

````python
Kmean.labels_
````

- Based on the inertia, 3 clusters are optimal because additional clusters do not improve its value.

Do the same but this time using the silhouette score. Compare both. Would you now reconsider your choice of the previous question?

````python
silhouette_s = []
for num_clusters in range(2, 10):
    kmeans_model = KMeans(n_clusters=num_clusters)
    kmeans_model.fit(df_data_kmeans)
    silhouette_sc = silhouette_score(df_data_kmeans, kmeans_model.labels_)
    silhouette_s.append(silhouette_sc)
````

````python
silhouette_s
````

````python
plt.figure(figsize=(10,3))
plt.plot(range(2,10),silhouette_s)
plt.scatter(range(2,10),silhouette_s)
# plt.scatter(3, inertia_list[3], marker="X", s=300, c="r")
plt.xlabel("Number of Clusters", size=13)
plt.ylabel("Silhoulette Score", size=13)
plt.title("Different Silhoulette Scores for Different Number of Clusters", size=17)
````

- The two measures indicate different number of clusters, 3 and 2 for inertia and silhouette, respectively. Based on this, it might be worth taking a closer look at the features and checking if there is indeed a need for 3 clusters. 

# Question 3

(22 points)

Mention one advantage and one disadvantage of using Principal Component Analysis (PCA).

- Advantages:
Removes correlated features because all the principal components are independent between each other. Improves algorithm Performance and it can also speed it up, since we are reducing the dimensions. Reduces overfitting because we use less variables in our models or algorithms. Moreover, it is easier to visualise because we have less dimensions.
  
  
- Disadvantages:
Independent variables become less interpretable and we can not interpret them as an independent variables like we could before. We must use standardised data before the PCA, because if we do not do it, the PCA will be biased towards variables with a high variance, which is misleading.

Reduce the dataset to two dimensions with PCA and make a clear plot. Make sure to label all the important information.

````python
df_PCA= df_data
````

````python
#Checking the type of data, so we can make a split.
df_PCA.dtypes
````

````python
#Split the data into numerical and categorical data.
df_features_num = df_PCA.loc[:,df_PCA.dtypes != object]
df_features_cat = df_PCA.loc[:, df_PCA.dtypes == object]
````

````python
#Standardize data using Standard Scaler, because we need to in order to run PCA.
scaler = StandardScaler()
features = df_features_num.columns
df_features_numstd_all = scaler.fit_transform(df_features_num)
df_features_numstd_all = pd.DataFrame(data=df_features_numstd_all, columns = features)
````

````python
#Dropping the target variable and Zip, because is not neccessary.
df_features_numstd= df_features_numstd_all.drop(columns=['ProbabilityDefault','Zip'])
````

````python
#PCA process and Plot of explained variance by two dimensions.
pca = PCA(n_components=2).fit(df_features_numstd)

plt.figure(figsize=(9,7))
plt.bar(range(pca.n_components), pca.explained_variance_ratio_)
plt.ylabel('Percentage of Variance Explained')
plt.xlabel('Principal Component')
plt.title('PCA Score Plot')
plt.show()
````

````python
pca.explained_variance_ratio_
````

````python
#Match index of the two data frames.
df_features_pca = pd.DataFrame(pca.transform(df_features_numstd))
df_features_pca.index=df_features_numstd.index
````

````python
#Rename the columns
df_features_pca.columns = [f'pca.{i}' for i in df_features_pca.columns]
````

````python
df_features_num1 = pd.concat([df_features_numstd, df_features_pca], axis=1)
````

This time reduce the dataset to two dimensions with t-SNE and make a clear plot.

````python
#Rename the columns
tsne_df.columns = [f'tsne.{i}' for i in tsne_df.columns]
````

````python
df_features_num1 = pd.concat([df_features_num1, tsne_df], axis=1)
````

````python
#Plot of PCA, colours =Probability of Default.
plt.figure(figsize=(15,10))
plt.scatter(x=df_features_num1['pca.0'], y=df_features_num1['pca.1'],
            c=df_features_num['ProbabilityDefault'],cmap='rainbow')
plt.ylabel('Principal Component 2')
plt.xlabel('Principal Component 1')
plt.title('PCA')
````

Add the two columns you made with PCA to the dataset. Then pick a Decision Tree model and train a good model to predict the Probability of Default (`ProbabilityDefault`). It is not necessary to tune the hyperparameters (so the default settings are fine). 

````python
#creating dummy variables for all categorical variables
dct_dummies = {}

for cat, sr in df_features_cat.iteritems():
    df_dummies = pd.get_dummies(sr.astype('category'))
    df_dummies.columns = [f'cat.{cat}.{c}' for c in df_dummies]
    dct_dummies[cat] = df_dummies.iloc[:,1:] # drop one to avoid multicollinearity
````

````python
df_features_dummies = pd.concat(dct_dummies.values(), axis=1)
df_features_dummies.index=df_features_num1.index
print(np.shape(df_features_dummies))
````

````python
df_const = pd.DataFrame(index=df_features_dummies.index, columns=['const'])
df_const.index=df_features_num1.index
df_const = df_const.fillna(1)
print(df_const)
````

````python
#creating a dataframe of features
df_features = pd.concat([
    df_const,
    df_features_dummies,
    df_features_num1, 
], axis=1)

print(df_features)
````

````python
y = df_features_numstd_all[['ProbabilityDefault']]
x = df_features.drop(columns=['tsne.0','tsne.1'])
````

````python
#changing the ProbabilityDefault variable to label, i.e. 0 or 1 variable
y['ProbabilityDefault'].values[y['ProbabilityDefault'].values>0] = 1
y['ProbabilityDefault'].values[y['ProbabilityDefault'].values<0] = 0
````

````python
y = y.astype(int)
````

````python
#Split the data into training and test set (both x and y variables)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=25)
````

````python
base_tree = tree.DecisionTreeClassifier(random_state=25)
````

````python
base_tree = base_tree.fit(x_train, y_train)
````

````python
y_predict_base = base_tree.predict(x_test)
````

````python
print(classification_report(y_test, y_predict_base))
````

````python
grid_search_cv.best_params_
````

````python
adv_tree=tree.DecisionTreeClassifier(random_state=25, max_depth=1, splitter="best" , min_samples_leaf=1)
````

````python
y_predict = adv_tree.predict(x_test)
````

````python
print(classification_report(y_test, y_predict))
````
